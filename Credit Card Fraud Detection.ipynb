```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sms
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card=pd.read_csv("/content/creditcard.csv")
```

```python
credit_card
```

```python
credit_card.info()
```

```python
credit_card.isnull().sum()
```

```python
credit_card['Class'].value_counts()
```

*Data is unbalanced*

0-> Legit transacation

1-> fraud transaction

***Seperating the data for analysis***

```python
legit=credit_card[credit_card.Class==0]
fraud=credit_card[credit_card.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card.groupby('Class').mean()
```

***Under-Sampling***

No. of fraud transactions->492

```python
legit_sample=legit.sample(n=492)
legit_sample
```

***Concatenating two dataframes***

```python
new_data=pd.concat([legit_sample , fraud],axis=0)
```

```python
new_data
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

***Splitting data into target and features***

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
print(X)
print(Y)
```

***Splitting data into testing and training***

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

***Modelling Training:-Logistic Regression***

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

***Model Evaluation***

***Accuracy Score***

```python
X_train_prediction=model.predict(X_train)
training_X_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy Score:-",training_X_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_X_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy Score:-",test_X_data_accuracy)
```
